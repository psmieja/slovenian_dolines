```python
import numpy as np
import matplotlib.pyplot as plt

from lidar_data_analyzer import LidarDataSet
```

```python
selected_range_points = [[1200, 4040],
                         [1586, 3777],
                         [2500, 3440],
                         [2852, 3100],
                         [3290, 2973],
                         [3790, 2382],
                         [3793, 2091],
                         [4470, 1700],
                         [6000, 270],
                         [6000, 0],
                         [4025, 0],
                         [2340, 1646],
                         [2280, 1930],
                         [1492, 2500],
                         [1193, 3120],
                         [465, 3700],
                         [310,4385]]
```

```python
data_segment_filenames = [[f"TM1_{x}_{y}" for y in range(80,84+1)] for x in range(442,447+1)]
lds = LidarDataSet.from_array_of_csv_files(data_segment_filenames,
                                           filename_prefix="./data/dataset2/",
                                           selected_range_points=selected_range_points)
lds.find_sinkholes()
lds.plot(centers=True, analyzed_region=True, verbose=False, fig_size=(12,9))
```

```python
lds.find_sinkhole_edges(verbose=False)
lds.plot(centers=False, edges=True, analyzed_region=True, verbose=False, fig_size=(12,9))
```

```python
lds.sinkholes.to_pickle("./program_data/sinkholes.pkl")
lds.sinkholes.to_csv("./program_data/sinkholes.csv")

with open("./program_data/dataset.npy", "wb") as f:
    np.save(f, lds.dataset)
```
